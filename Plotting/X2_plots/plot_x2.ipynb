```python
#load a file
file_25="//global/scratch/users/jennaisrael/run_schism/run_25/x2/x2_ts_bay_sac_missing_data.csv"
file_26="//global/scratch/users/jennaisrael/run_schism/run_26/x2/x2_ts_bay_sac.csv"
file_27="//global/scratch/users/jennaisrael/run_schism/run_27/x2/x2_ts_bay_sac.csv"
file_28="//global/scratch/users/jennaisrael/run_schism/run_28/x2/x2_ts_bay_sac_missing_data.csv"
data_25=pd.read_csv(file_25,names=['Date','X2'],skiprows=[0])
data_26=pd.read_csv(file_26,names=['Date','X2'],skiprows=[0])
data_27=pd.read_csv(file_27,names=['Date','X2'],skiprows=[0])
data_28=pd.read_csv(file_28,names=['Date','X2'],skiprows=[0])
data_28
# data = pd.read_csv(file, sep='\s+', names=columns) #, decode_times=False)
# data = data.apply(pd.to_numeric)

 
```

```python
# Add in the number of seconds elapsed @ our chosen timestep
data_25['Date']=pd.to_datetime(data_25.Date, format = "%Y-%m-%d")
data_26['Date']=pd.to_datetime(data_26.Date, format = "%Y-%m-%d")
data_27['Date']=pd.to_datetime(data_27.Date, format = "%Y-%m-%d")
data_28['Date']=pd.to_datetime(data_28.Date, format = "%Y-%m-%d")
```

```python
fig = plt.figure(figsize=(20,6))
mpl.rcParams.update({'font.size': 22})
ax = plt.gca()

ax.plot(data_26.Date[90:], data_26.X2[90:], label = 'Historical', linewidth=2, alpha =0.7)
ax.plot(data_25.Date[90:], data_25.X2[90:], label = '0.8 m RSLR', linewidth=2, alpha =0.7)  
ax.plot(data_27.Date[90:], data_27.X2[90:], label = '1.2 m RSLR', linewidth=2, alpha =0.7)
ax.plot(data_28.Date[90:], data_28.X2[90:], label = '+2 C', linewidth=2, alpha =0.7)

leg = ax.legend()
print(ax)
ax.set_xlim(data_27.Date.iloc[90], data_27.Date.iloc[-1])
ax.set_ylabel('Daily average X2 [km]')
# ax.grid(which='major', color='b', linestyle='-')
#ax.set_title("X2 timeseries")
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%d-%Y'))
plt.show()
# ax.grid()
```
