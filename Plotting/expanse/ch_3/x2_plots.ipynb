```python
c=81 #compliance value
sdate=pd.to_datetime("2018-10-1")
edate=pd.to_datetime("2019-3-31")
months=np.unique(base1[sdate:edate].index.month.values)
daysabovec_1=np.empty(months.shape)
daysabovec_2=np.empty(months.shape)
daysabovec_3=np.empty(months.shape)
# make a df to write to csv
stadays1=np.array([])
stadays2=np.array([])
stadays3=np.array([])
comp1=base1[sdate:edate].where(base1>c)
comp2=dcp1[sdate:edate].where(dcp1>c)
comp3=cor1[sdate:edate].where(cor1>c)

print('baseline has '+str(len(comp1.dropna()))+' days above compliance')
print('dcp has '+str(len(comp2.dropna()))+' days above compliance')
print('cor has '+str(len(comp3.dropna()))+' days above compliance')

for m in months:
    #print(stadays)
    days1=len(comp1[comp1.index.month==m].dropna())
    stadays1=np.append(stadays1,days1)
    days2=len(comp2[comp2.index.month==m].dropna())
    stadays2=np.append(stadays2,days2)
    days3=len(comp3[comp3.index.month==m].dropna())
    stadays3=np.append(stadays3,days3)
```

```python
# add a row for x2 compliance
df_compliance1.loc[-1]=stadays1
df_compliance2.loc[-1]=stadays2
df_compliance3.loc[-1]=stadays3
df_compliance1
```
