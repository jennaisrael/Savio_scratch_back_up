```python
# make stacked plots 
fig, ax = plt.subplots(4,figsize=(20,12),sharex=True,layout='constrained')
mpl.rcParams.update({'font.size': 22})

ax[0].plot(base1.x2,label='Baseline', linewidth=3,color=cmocean.cm.tarn(1/6))
ax[0].plot(dcp1.x2,label='DCP', linewidth=3,color=cmocean.cm.tarn(10/12))
ax[0].plot(cor1.x2,label='CoR', linewidth=3,color=cmocean.cm.tarn(2/6))
ax[0].set_title('sac')
ax[0].set_ylabel('X2 [km]')
ax[0].set_xlim(pd.to_datetime("2018-10-1"),pd.to_datetime("2019-9-30"))
#ax[0].set_ylim(0,5000)
ax[0].legend()

ax[1].plot(base2.x2,label='Baseline', linewidth=3,color=cmocean.cm.tarn(1/6))
ax[1].plot(dcp2.x2,label='DCP', linewidth=3,color=cmocean.cm.tarn(10/12))
ax[1].plot(cor2.x2,label='CoR', linewidth=3,color=cmocean.cm.tarn(2/6))
ax[1].set_title('sjr')
ax[1].set_ylabel('X2 [km]')
ax[1].set_xlim(pd.to_datetime("2018-10-1"),pd.to_datetime("2019-9-30"))


ax[2].plot(base3.x2,label='Baseline', linewidth=3,color=cmocean.cm.tarn(1/6))
ax[2].plot(dcp3.x2,label='DCP', linewidth=3,color=cmocean.cm.tarn(10/12))
ax[2].plot(cor3.x2,label='CoR', linewidth=3,color=cmocean.cm.tarn(2/6))
ax[2].set_title('nysjr')
ax[2].set_ylabel('X2 [km]')
ax[2].set_xlim(pd.to_datetime("2018-10-1"),pd.to_datetime("2019-9-30"))

ax[3].plot(base4.x2,label='Baseline', linewidth=3,color=cmocean.cm.tarn(1/6))
ax[3].plot(dcp4.x2,label='DCP', linewidth=3,color=cmocean.cm.tarn(10/12))
ax[3].plot(cor4.x2,label='CoR', linewidth=3,color=cmocean.cm.tarn(2/6))
ax[3].set_title('mzm')
ax[3].set_ylabel('X2 [km]')
ax[3].set_xlim(pd.to_datetime("2018-10-1"),pd.to_datetime("2019-9-30"))


#fig.savefig('/home/jisrael/Savio_scratch_back_up/Plotting/expanse/ch_3_figures/X2_allpaths_scenarios.png')
```

```python
#just plot sac path
fig, ax = plt.subplots(figsize=(12,4),sharex=True,layout='constrained')
mpl.rcParams.update({'font.size': 22})

ax.plot(base1.x2,label='Baseline', linewidth=3,color=cmocean.cm.tarn(1/6))
ax.plot(dcp1.x2,label='DCP', linewidth=3,color=cmocean.cm.tarn(10/12),linestyle='--')
ax.plot(cor1.x2,label='CoR', linewidth=3,color=cmocean.cm.tarn(2/6),linestyle=(0,(1,1)))
#ax.hlines(y=80, xmin=pd.to_datetime("2018-2-1"), xmax=pd.to_datetime("2019-9-30"),color='k',linestyle='--',label='compliance')
ax.set_ylabel('X2 [km]')
ax.set_xlim(pd.to_datetime("2018-10-1"),pd.to_datetime("2019-1-31"))
ax.legend(frameon=False)
ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(ax.xaxis.get_major_locator()))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%b'))
    # Rotates and right-aligns the x labels so they don't crowd each other.
for label in ax.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')
plt.show()
#fig.savefig('/home/jisrael/Savio_scratch_back_up/Plotting/expanse/ch_3_figures/X2_sac_scenarios_endfeb_no_comp.png')
```
