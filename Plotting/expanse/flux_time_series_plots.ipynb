```python
timedelta1 = [pd.Timedelta(i, unit = 's') for i in flux1.time]
timestamp1 = [pd.to_datetime("01-01-2017 00:00") + i for i in timedelta1]
flux1["time"] = timestamp1

timedelta2 = [pd.Timedelta(i, unit = 's') for i in flux2.time]
timestamp2 = [pd.to_datetime("01-01-2017 00:00") + i for i in timedelta2]
flux2["time"] = timestamp2

timedelta3 = [pd.Timedelta(i, unit = 's') for i in flux3.time]
timestamp3 = [pd.to_datetime("01-01-2017 00:00") + i for i in timedelta3]
flux3["time"] = timestamp3
```

```python
#too many datapoints to plot the raw data so try taking a daily average 
#need to set datetime as index first
flux1_dated=flux1.set_index(pd.DatetimeIndex(flux1.time))
flux2_dated=flux2.set_index(pd.DatetimeIndex(flux2.time))
flux3_dated=flux3.set_index(pd.DatetimeIndex(flux3.time))
print(flux3_dated)
```

```python
flux1_daily=flux1_dated.resample('D').mean()
flux2_daily=flux2_dated.resample('D').mean()
flux3_daily=flux3_dated.resample('D').mean()
print(flux3_daily)
```

```python
fig = plt.figure(figsize=(16,6))
ax1 = fig.gca()

ax1.plot(flux1_daily.swp+flux1_daily.cvp,label='Historical',linewidth=5)
ax1.plot(flux2_daily.swp+flux2_daily.cvp,label='Decr Exp',linestyle=':',linewidth=5)
ax1.plot(flux3_daily.swp+flux3_daily.cvp,label='Incr Inf',linestyle='--',linewidth=5)
ax2 = ax1.twinx()
ax2.plot(flux1_daily.sac,label='Historical',linewidth=5)
ax2.plot(flux2_daily.sac,label='Decr Exp',linestyle=':',linewidth=5)
ax2.plot(flux3_daily.sac,label='Incr Inf',linestyle='--',linewidth=5)

ax1.set_xlim(flux1_daily.time[542],flux1_daily.time[572])
ax1.set_ylabel('SWP + CVP Pumping $(m^3/s)$',fontsize=20)
ax1.set_ylim(0,200)
ax2.set_ylabel('Sacramento River Flow $(m^3/s)$',fontsize=20)
ax2.set_ylim(-400,-200)
leg = ax1.legend()
plt.show()
```
