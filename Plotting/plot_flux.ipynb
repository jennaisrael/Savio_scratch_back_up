```python
import numpy as np
import pandas as pd
import sys
import os, sys
import matplotlib.pyplot as plt
import xarray as xr
```

```python
flux2=pd.read_table('/global/scratch/users/jennaisrael/BayDeltaSCHISM/data/time_history/flux.th',delimiter=' ',header=0, names=['datetime','coyote','ccc_rock','ccc_old','swp','cvp','sjr','calaveras','east','american','sac','yolo_toedrain','yolo','northbay','napa','ccc_victoria'],parse_dates=True)
```

```python
flux2['swp'].plot()
```

```python
#too many datapoints to plot the raw data so try taking a daily average 
#need to set datetime as index first
flux_dated=flux2.set_index(pd.DatetimeIndex(flux2.datetime))
print(flux_dated)
```

```python
flux_daily=flux_dated.resample('D').mean()
print(flux_daily)
```

```python
ax = plt.gca()
ax.plot(flux_daily.swp,label='swp')
ax.plot(flux_daily.cvp,label='cvp')
```

```python
fig = plt.figure(figsize=(16,6))
ax = fig.gca()

ax.plot(flux_daily.swp,label='swp')
ax.plot(flux_daily.cvp,label='cvp')
ax.set_xlim('2013-01-01','2021-01-01')
ax.set_ylabel('Pumping Flux (m^3/s)')
leg = ax.legend()
plt.show()
```
