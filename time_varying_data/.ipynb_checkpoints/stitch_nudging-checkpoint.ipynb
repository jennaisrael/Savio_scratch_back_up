For a run with 2 datatypes for nudging (chang in data source/ file path), concatenate the TEM and SAL netcdfs together before starting run

```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt 
#import netcdf4
```

```python
uv3d_24=xr.open_dataset("/global/scratch/users/jennaisrael/run_schism/run_24/uv3D.th.nc")
uv3d_24
```

```python
#use same index as below to trim and reset the time indices of this file
#238 days *3600 s/h *24 h/day 
sec_elapsed=238*3600*24
uv3d_24_masked = uv3d_24.where(uv3d_24.time > sec_elapsed, drop=True)
uv3d_24_masked
```

```python
offset=int(uv3d_24_masked.time[0]) #the first value is 0 so the second entry is the frequency
print(offset)
offset_v=np.ones_like(np.array(uv3d_24_masked.time)) * offset
uv3d_24_masked_reset=uv3d_24_masked.copy()
uv3d_24_masked_reset["time"]=uv3d_24_masked.time-offset_v
print(uv3d_24_masked_reset)
```

```python
#write this one to a file for run_26
uv3d_24_masked_reset.to_netcdf(path="/global/scratch/users/jennaisrael/run_schism/run_26/uv3D.th.nc")
```

#For temperature and Salinity

```python
tem_24=xr.open_dataset("/global/scratch/users/jennaisrael/run_schism/run_24/TEM_nu_roms.nc")
sal_24=xr.open_dataset("/global/scratch/users/jennaisrael/run_schism/run_24/SAL_nu_roms.nc")
sal_24
```

```python
tem_masked = tem_24.where(tem_24.time > 20563200, drop=True)
tem_masked
```

```python
offset=int(tem_masked.time[0]) #the first value is 0 so the second entry is the frequency
print(offset)
offset_v=np.ones_like(np.array(tem_masked.time)) * offset
#print(offset_v)
```

```python
tem_masked_reset=tem_masked.copy()
tem_masked_reset["time"]=tem_masked.time-offset_v
print(tem_masked_reset)
```

```python
#repeat for salinity
sal_masked = sal_24.where(sal_24.time > 20563200, drop=True)
sal_masked_reset=sal_masked.copy()
sal_masked_reset["time"]=sal_masked.time-offset_v
print(sal_masked_reset)
```

```python
#write the files to the run_25 directory
tem_masked_reset.to_netcdf(path="/global/scratch/users/jennaisrael/run_schism/run_25/TEM_nu_roms.nc")
sal_masked_reset.to_netcdf(path="/global/scratch/users/jennaisrael/run_schism/run_25/SAL_nu_roms.nc")
```

```python
#are the gr3 files the same
tem_1=xr.open_dataset("/global/scratch/users/jennaisrael/schism_scripts_scratch/nudging4baroclinic/TEM_nu_roms_1.nc")
tem_2=xr.open_dataset("/global/scratch/users/jennaisrael/schism_scripts_scratch/nudging4baroclinic/TEM_nu_roms_2.nc")
print(tem_1)
```

```python
print(tem_2)
```

```python
#I think need to update the time of the second nc first
#add the last time value from the first file plus the delta between time points to the second time vector
offset=int(tem_1.time[-1]+tem_2.time[1]) #the first value is 0 so the second entry is the frequency
#print(offset)
offset_v=np.ones_like(np.array(tem_2.time)) * offset
print(offset_v)
tem_2_update=tem_2.copy()
tem_2_update["time"]=tem_2.time+offset_v
print(tem_2_update)
```

```python
tem_total=xr.concat([tem_1,tem_2_update],dim="time") 
```
